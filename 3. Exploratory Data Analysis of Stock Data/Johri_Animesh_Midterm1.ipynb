# Midterm1 - Exploratory Data Analysis of Stock Data

##### Description about data: 
7 Columns in all the CSVs having stock data of 31 companies: 
1. Date:It is the date of the trading day. 
2. Open It is the price of the stock at the beginning of the trading day (it need not be the closing price of the previous trading day). 
3. high: It is the highest price of the stock on that trading day. 
4. low: It is the lowest price of the stock on that trading day.  
5. close: It is the price of the stock at closing time. 
6. Volume: It indicates how many stocks were traded on that day.
7. Name: Name of the Company

##### CSV File details: 
1.	All CSV file's dates starting from 1st Jan 2006 till the day before 1st Jan 2018.
2.	The first row starts from 3rd Jan 2006(Tuesday) because 1st working day of 2006 is 2nd January(Monday), Data for 3rd January was collected on 2nd January. 
3.	The last row has a date of 29th Dec 2017(Friday)- Last working day of the year. 
4.	For All Stocks (Jan 2017 – Jan 2018), the first date is 3rd Jan 2017(Tuesday) because 1st working day of 2017 is 2nd January(Monday), Data for 3rd January was collected on 2nd January.

```python
# Creating a list of the ticker symbols (as strings) in alphabetical order. Calling this list: tickers
tickers = ['AMZN', 'GOOGL']
```

#### Concatenating the dataframes together to a single data frame called company_stocks. Set the keys argument equal to the tickers list. 

```python
#Concatenating AMZN and GOOGLE on Date and assinging the tickers as key
company_stocks = pd.concat([AMZN, GOOGL], axis=1, keys=tickers)
```

```python
company_stocks.columns.names = ['Company Ticker','Stock Info']
company_stocks.head()
```

#### Creating dataframe that will contain the returns for each company's stock. It will compare the value of stock at closing time on the previous day.

```python
returns = pd.DataFrame()

for i in tickers:
    returns[i + ' Return'] = company_stocks.xs((i, 'Close'), axis=1).pct_change()

returns.dropna(inplace=True)
```
